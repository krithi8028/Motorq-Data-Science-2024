```python
def check_vehicle_id_consistency(telemetry_ids, mapping_ids):
    telemetry_set = set(telemetry_ids)
    mapping_set = set(mapping_ids)
    if telemetry_set == mapping_set:
        print("Vehicle IDs are consistent across files")
    else:
        print("Inconsistency in vehicle IDs:")
        print(f"IDs in telemetry but not in mapping: {telemetry_set - mapping_set}")
        print(f"IDs in mapping but not in telemetry: {mapping_set - telemetry_set}")
check_vehicle_id_consistency(teledata['VEHICLE_ID'], vehicle_pnid_mapping['ID'])
```

```python
def check_battery_level_range(df): #to check battery level range
    invalid = df[(df['EV_BATTERY_LEVEL'] < 0) | (df['EV_BATTERY_LEVEL'] > 100)]
    if len(invalid) > 0:
        print(f"Found {len(invalid)} records with invalid battery level")
    else:
        print("All battery levels are within the expected range (0-100%)")
check_battery_level_range(teledata)
```

```python
def check_odometer_consistency(df):
    df_sorted = df.sort_values(['VEHICLE_ID', 'TIMESTAMP'])
    inconsistent = df_sorted.groupby('VEHICLE_ID').filter(lambda x: (x['ODOMETER'].diff() < 0).any())
    if len(inconsistent) > 0:
        print(f"Found {len(inconsistent)} records with inconsistent odometer values")
    else:
        print("All odometer values are consistent (always increasing)")
check_odometer_consistency(teledata)
```

```python
charge_state_soc = triggers_soc[triggers_soc['name'] == 'CHARGE_STATE'][['VEHICLE_ID', 'TIMESTAMP', 'val']]

# Initialize a new column for charge level
filtered_all_event_df['CHARGE_LEVEL'] = None

# Define time window in seconds
time_window = timedelta(seconds=300)

# Function to find the closest battery level within the time window
def find_closest_battery_level(event_row, telemetry_data, charge_state_soc):
    vehicle_id = event_row['VEHICLE_ID']
    event_time = event_row['TIMESTAMP']

    # Filter data for the specific vehicle
    telemetry_vehicle = telemetry_data[telemetry_data['VEHICLE_ID'] == vehicle_id]
    charge_vehicle = charge_state_soc[charge_state_soc['VEHICLE_ID'] == vehicle_id]

    # Find the closest telemetry data
    telemetry_close = telemetry_vehicle.loc[
        (telemetry_vehicle['TIMESTAMP'] >= event_time - time_window) &
        (telemetry_vehicle['TIMESTAMP'] <= event_time + time_window)
    ]

    # Find the closest charge state data
    charge_close = charge_vehicle.loc[
        (charge_vehicle['TIMESTAMP'] >= event_time - time_window) &
        (charge_vehicle['TIMESTAMP'] <= event_time + time_window)
    ]

    # Get the closest timestamp and corresponding battery level
    if not telemetry_close.empty:
        closest_telemetry = telemetry_close.iloc[(telemetry_close['TIMESTAMP'] - event_time).abs().argsort()[:1]]
    else:
        closest_telemetry = pd.DataFrame()

    if not charge_close.empty:
        closest_charge = charge_close.iloc[(charge_close['TIMESTAMP'] - event_time).abs().argsort()[:1]]
    else:
        closest_charge = pd.DataFrame()

    # Determine the closest overall
    if not closest_telemetry.empty and not closest_charge.empty:
        if abs(closest_telemetry['TIMESTAMP'].iloc[0] - event_time) <= abs(closest_charge['TIMESTAMP'].iloc[0] - event_time):
            return closest_telemetry['EV_BATTERY_LEVEL'].iloc[0]
        else:
            return closest_charge['val'].iloc[0]
    elif not closest_telemetry.empty:
        return closest_telemetry['EV_BATTERY_LEVEL'].iloc[0]
    elif not closest_charge.empty:
        return closest_charge['val'].iloc[0]
    else:
        return None

# Apply the function to find the closest charge level for each event
filtered_all_event_df['CHARGE_LEVEL'] = filtered_all_event_df.apply(
    find_closest_battery_level, axis=1, args=(telemetry_data, charge_state_soc)
)

# Final dataset
final_dataset = filtered_all_event_df[['VEHICLE_ID', 'EVENT_NAME', 'TIMESTAMP', 'CHARGE_LEVEL']]
final_dataset.to_csv('final_dataset.csv', index=False)

# Output the final dataset
print(final_dataset)
```
